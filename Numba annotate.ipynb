```python
import numpy as np
from numba import jit

class clazz(object):

    def __init__(self, arr):
        self._arr = arr

    @property
    def arr(self):
        return self._arr

    @arr.setter
    def arr(self, value):
        self._arr = value


def qux(a, c):
    """Example function to test numba annotate can be improved.
    
    We want syntax highlighting, and stuff.
    """
    c.arr += 12

    for i in range(len(a)):
        a[i] = np.sqrt(i) + 7
        if i % 8 + 1 > 4:
            a[i] -= np.pi


    return a

A = np.arange(100.)
c1 = clazz(11)
c2 = clazz(11)
```

```python
jitqux = jit(qux)
```

```python
jitqux(A, c2);
```
