Python Implementation---------------------

```python
def pyfilter2d(image, filt):    M = len(image)    N = len(image[0])    Mf = len(filt)    Nf = len(filt[0])    Mf2 = Mf // 2    Nf2 = Nf // 2    result = []    for i in range(M):        inner = []        for j in range(N):            num = 0.0            if i > Mf2-1 and i < M-Mf2 and j > Nf2-1 and j < N-Nf2:                for ii in range(Mf):                    for jj in range(Nf):                        num += filt[Mf-1-ii][Nf-1-jj] * image[i-Mf2+ii][j-Mf2+jj]            inner.append(num)        result.append(inner)    return result
```

```python
out2 = pyfilter2d(image.tolist(), filt)
```

```python
imshow(out2)
```

NumPy Implemenation (vectorized)--------------------------------

```python
def numfilter2d(image, filt):    filt = np.asarray(filt)    Mf, Nf = filt.shape    Mf2 = Mf//2    Nf2 = Nf//2    result = np.zeros_like(image)    M, N = image.shape    for ii in range(Mf):        for jj in range(Nf):            result[Mf2:-Mf2,Nf2:-Nf2] += filt[ii,jj] * image[-1+Mf-ii:M-ii,-1+Nf-jj:N-jj]    return result
```

```python
out3 = numfilter2d(image, filt)
```

```python
imshow(out3)
```

Weave Implemenation-------------------

```python
import scipy.weave as weavedef weave_filter2d(image, filt):    filt = np.asarray(filt, dtype=image.dtype)    res = np.zeros_like(image)    code = """    int M, N, Mf, Nf;    int i, j, ii, jj;    double num;    Mf = Nfilt[0];    Nf = Nfilt[1];    M = Nimage[0];    N = Nimage[1];    for (i=2; i<M-2; i++) {        for (j=2; j<N-2; j++) {            num = 0.0;            for (ii=0; ii<Mf; ii++) {                for (jj=0; jj<Nf; jj++) {                    num += FILT2(Mf-1-ii, Nf-1-jj)*IMAGE2(i-2+ii, j-2+jj);                }            }            RES2(i, j) = num;        }    }    """    weave.inline(code, arg_names=['image', 'filt', 'res'])    return res
```

```python
out3 = weave_filter2d(image, filt)
```
